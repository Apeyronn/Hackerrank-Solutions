```python
def minimumBribes(q):
    n = len(q)
    bribes = 0
    
    for i in range(n-1, -1, -1):
        # Check if person i bribed more than 2 people
        if q[i] - (i+1) > 2:
            print("Too chaotic")
            return
        
        # Count the number of people that person i bribed
        for j in range(max(0, q[i]-2), i):
            if q[j] > q[i]:
                bribes += 1
        
    print(bribes)

minimumBribes([1, 2, 4 ,3])
```
