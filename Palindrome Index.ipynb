```python
def isPalindrome(word):
    reverse = word[::-1]
    return True if reverse == word else False

def palindromeIndex(s):
    # Write your code here
    if isPalindrome(s):
        return -1
    for i in range(len(s)//2):
        if s[i] != s[len(s)-1-i]:
            new_s = s[:i]+s[i+1:]
            if isPalindrome(new_s):
                return i
            else:
                return len(s) - i - 1
        
    return -1
print(palindromeIndex('aaba'))
```
