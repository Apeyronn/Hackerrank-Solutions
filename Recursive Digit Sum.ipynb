```python
def superDigit(n, k): #1,3 n:9875 k:1
    # Calculate the super digit of a single integer
    def calcSuperDigit(x): #4,5 x:9875 8 x:9875
        if x < 10: #
            return x #
        else: #
            return calcSuperDigit(sum(int(d) for d in str(x))) #10 11 d:9,8,7,5

    # Calculate the super digit of n
    p = int(n) * k #6 p:9875
    return calcSuperDigit(p) # 7
a=superDigit(9875,1) #2

print(a)
```
