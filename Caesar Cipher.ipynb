```python
def caesarCipher(s, k):
    lower_case = 'abcdefghijklmnopqrstuvwxyz'
    upper_case = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

    ans = []

    for c in s:
        if c in lower_case:
            index = lower_case.index(c)
            ans.append(lower_case[(index + k) % 26])
        elif c in upper_case:
            index = upper_case.index(c)
            ans.append(upper_case[(index + k) % 26])
        else:
            ans.append(c)

    return "".join(ans)


print(caesarCipher('merhaba', +1))
```
